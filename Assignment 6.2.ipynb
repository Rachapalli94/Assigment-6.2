```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
titanic=pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv')
titanic.head()
```

```python
#pie chart plot for ratio of male and female
titanic['sex1'],_=pd.factorize(titanic['sex'])
```

```python
m_f=titanic.sex1
m_f.astype(bool).sum()
m=844
f=1309-844
labels='males','females'
sizes=[844,465]
colors=['blue','green']
explode=(0.1,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True,startangle=140)
plt.axis('equal')
plt.show()
```

```python
#scatter plot for fare and age of male and female
plt.scatter(titanic.fare,titanic.age,c=titanic.sex1)
plt.title('fare vs age')
plt.ylabel('age')
plt.xlabel('fare')
```
